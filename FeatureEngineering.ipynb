### Improting libraries

```python
import numpy as np
import pandas as pd
```

### Loading Data

```python
dataset = pd.read_csv("Dataset/BankCustomers.csv")
```

### Feature Engineering

```python
dataset.drop(columns=["RowNumber", "CustomerId", "Surname"], inplace=True)
```

```python
categorical_data = dataset.select_dtypes("object")
cat_data = pd.get_dummies(categorical_data, drop_first=True)
cat_data.head()
```

```python
dataset.drop(columns=categorical_data, inplace=True)
dataset.head()
```

```python
scaling_data = dataset[["CreditScore", "Age", "Balance", "EstimatedSalary"]]
dataset.drop(columns = scaling_data.columns, inplace=True)
scaling_data.head()
```

```python
dataset.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaling = StandardScaler()
data1 = scaling.fit_transform(scaling_data)
```

```python
data1.shape
```

```python
final_data = pd.concat([pd.DataFrame(data1), dataset, cat_data], axis=1)
final_data.head()
```

```python
final_data.to_csv("Dataset/FeatureEnginnered1.csv")
```
